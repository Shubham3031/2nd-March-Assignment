```python
# 1.Ans.

Matplotlib is a data visualization library for Python programming language that is used to 
create static, animated, and interactive visualizations in Python. It is an open-source and
easy-to-use library that provides a wide variety of customizable plotting options.

Matplotlib is used to create visualizations of data for various purposes like scientific research, 
data analysis, and presentation of results. It is widely used in the fields of data science, 
statistics, and machine learning.

Here are five plots that can be plotted using the Pyplot module of Matplotlib:

Line Plot: A line plot is a basic plot that displays the relationship between two variables on the 
x and y-axis.

Scatter Plot: A scatter plot is a plot that shows the relationship between two variables as dots 
on a two-dimensional plane.

Bar Chart: A bar chart is a chart that represents data using rectangular bars. The length of each
bar is proportional to the value of the item it represents.

Histogram: A histogram is a graphical representation of the distribution of a dataset. It is a way 
to visualize the shape of a distribution.

Pie Chart: A pie chart is a chart that shows the proportion of each category in a dataset as a slice
of a circle.
```

```python
# 3.Ans.
"""
The subplot() function is used in Matplotlib, a popular Python
library for data visualization, to create multiple plots in a single figure.

"""
x = np.array([0, 1, 2, 3, 4, 5]) and  y = np.array([0, 100, 200, 300, 400, 500])
x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70]) 
x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

```python
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70]) 

x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])

fig, axs = plt.subplots(2, 2, figsize=(10, 8))


axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Plot 1')


axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Plot 2')


axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Plot 3')


axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Plot 4')


fig.suptitle('Four Line Plots')

plt.show()
```

```python
# 4.Ans.

Bar plot is a type of graph that represents categorical data with rectangular bars.

Bar plots can be used to show the distribution of data, compare different groups or
categories, and visualize changes over time. They can be used for both quantitative and qualitative data.
```

```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("bar plot")
```

```python
plt.barh(company,profit)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Horizontal bar plot")
```

```python
# 5.Ans.

A box plot, also known as a box-and-whisker plot, is a type of graphical representation
of a set of numerical data through their quartiles. Box plots are used to summarize the
distribution of the data by showing the median, quartiles, and outliers of the data.

Box plots are useful because they provide a concise summary of the distribution of 
a set of data, and they allow for easy comparison of multiple datasets. They can also 
be used to identify any outliers or extreme values in the data.
```

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

fig, ax = plt.subplots()

ax.boxplot([box1, box2])

ax.set_title("Box Plot")
ax.set_xticklabels(["Box 1", "Box 2"])
ax.set_ylabel("Values")

plt.show()

```
